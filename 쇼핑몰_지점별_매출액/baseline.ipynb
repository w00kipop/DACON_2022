데이터 전처리

```python
# 결측치 제거
train = train.fillna(0)
```

```python
train.info()
```

```python
# 일/월/년도 형식의 문자를 숫자로 변환해야한다. (이번에는 월만 사용해보도록 하겠습니다.)
def get_month(date):
    month = date[3:5]
    month = int(month)

    return month


train['Month'] = train['Date'].apply(get_month)
```

```python
train
```

```python
# Is Holiday의 (True, False) -> (1, 0)
def holiday_to_number(isholiday):
    if isholiday == True:
        number = 1
    else:
        number = 0
    
    return number

train['NumberHoliday'] = train['IsHoliday'].apply(holiday_to_number)
```

```python
train
```

```python
# test에도 위에 사항들을 적용
test = test.fillna(0)
test['Month'] = test['Date'].apply(get_month)
test['NumberHoliday'] = test['IsHoliday'].apply(holiday_to_number)
```

```python
test
```

모델링

```python
# 이번 프로젝트에서는 선형 회귀 모델을 사용합니다.
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# 분석할 의미가 없는 컬럼 제거
train = train.drop(columns=['id'])
test = test.drop(columns=['id'])

# 전처리 하기 전 컬럼들 제거
train = train.drop(columns=['Date', 'IsHoliday'])
test = test.drop(columns=['Date', 'IsHoliday'])

# 학습할 정보와 예측하고자 하는 정보 분리
x_train = train.drop(columns=['Weekly_Sales'])
y_train = train[['Weekly_Sales']]
```

```python
x_train
```

```python
y_train
```

```python
model.fit(x_train, y_train) # 모델 학습
```

```python
# sklearn으로 선언된 모델은 predict() 메서드를 통해 간단하게 결과를 예측할 수 있다.
prediction = model.predict(test)
print('-------------------예측된 상위 10개의 데이터 확인-------------------')
print(prediction[:10])
```

제출하기

```python
sample_submission['Weekly_Sales'] = prediction

sample_submission.head()
```

```python
sample_submission.to_csv('submission.csv', index=False)
```
